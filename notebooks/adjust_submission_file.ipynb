# Load test

```python
submission_dict_with_dummy = {}
submission_dict_with_dummy["modality"] = "RGB"

present_keys = []
missing_keys = []


with open("/Users/artemmerinov/data//holoassist/test_action_clips.txt", 'r') as f:
    for line in f:
        
        record = line.strip().split('_')
        video_name = "_".join(record[:-2])
        start = record[-2]
        end = record[-1]
        key = f"{video_name}_{float(start):.3f}_{float(end):.3f}"

        if key in submission_dict.keys():
            present_keys.append(key)
            submission_dict_with_dummy[key] = submission_dict[key]

        elif key not in submission_dict.keys():
            missing_keys.append(key)
            submission_dict_with_dummy[key] = [0,0,0,0,0]

        else:
            print(key)
```

```python
def find_duplicates(input_list):
    count_dict = {}
    duplicates = []

    for item in input_list:
        if item in count_dict:
            count_dict[item] += 1
        else:
            count_dict[item] = 1

    for item, count in count_dict.items():
        if count > 1:
            duplicates.append(item)

    return duplicates

duplicates = find_duplicates(present_keys)
duplicates
```

# 1. use .3f format

```python
all_keys = []

with open("/Users/artemmerinov/data//holoassist/test_action_clips.txt", 'r') as f:
    for line in f:

        record = line.strip().split('_')
        video_name = "_".join(record[:-2])
        start = record[-2]
        end = record[-1]

        key = f"{video_name}_{float(start):.3f}_{float(end):.3f}"
        all_keys.append(key)
```

```python
for k in all_keys:
    if k not in submission_dict:
        submission_dict[k] = [0,0,0,0,0]
```

# 2. Trim zeros

```python
trimmed_submission_dict = {}
trimmed_submission_dict["modality"] = "RGB"

for k,v in submission_dict.items():
    if k != "modality":
        record = k.strip().split('_')
        video_name = "_".join(record[:-2])
        start = record[-2]
        end = record[-1]
        trimmed_key = f"{video_name}_{float(start)}_{float(end)}" # trimmed key
        trimmed_submission_dict[trimmed_key] = v
```
